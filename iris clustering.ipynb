```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
```

```python
X_names = iris['feature_names']
type(X_names)
```

```python
sns.pairplot(data=pd.DataFrame(X, columns= X_names))
plt.show()
```

```python
inertia=[]

for i in range(1,6):
    kmeans = KMeans(n_clusters= i)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    
```

```python

plt.plot(range(1,6),inertia)
plt.xticks(labels= range(1,6),ticks=range(1,6))
plt.show()
```

```python
best_model = KMeans(n_clusters=3)
best_model.fit(X)
```

```python
labels= best_model.fit_predict(X)
labels
```

```python
X.shape
```

```python
centroids= best_model.cluster_centers_
```

```python
for i in range(0,4):
    for j in range(0,4):
        plt.scatter(X[:,i],X[:,j],c=labels)
        plt.scatter(centroids[:,i],centroids[:,j],s=100,marker= 'D')
        plt.xlabel(f'{X_names[i]}')
        plt.ylabel(f'{X_names[j]}')
        
        plt.show()
```
